```python
import pandas as pd
```

```python
df = pd.read_csv('Amazon Sales data.csv') # load the data set 
```

```python
print(df.head(5)) # display the first few rows
```

```python
# Data Cleaning and Transformation

```

```python
# Check the missing values
print(df.isnull().sum())
```

```python
df.replace([float('inf'),float('-inf')],float('nan'),inplace=True)
```

```python
# Drop rows with missing values

df.dropna(inplace=True)
```

```python
df
```

```python
# convert order_date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df
```

```python
# Extract month and year from 'order_date'
df['month']=df['Order Date'].dt.month
```

```python
df
```

```python
# Extract Year
df['year']=df['Order Date'].dt.year
```

```python
print(df.head(5))
df
```

```python
# Save the new df file
df.to_csv('Amazon_Analysis_file.csv',index=False)
```

```python
# calculate month-wise sales
month_wise_sales = df.groupby('month')['Sales Channel'].sum().reset_index()
```

```python
month_wise_sales
```

```python
month_wise_sales.to_csv('month_wise_data.csv',index=False)
```

```python
# Calculate year-wise sales
year_wise_sales = df.groupby('year')['Sales Channel'].sum().reset_index()
```

```python
year_wise_sales
```

```python
year_wise_sales.to_csv('Year_data.csv',index=False)
```

```python
yearly_month_wise_sales = df.groupby(['year','month'])['Sales Channel'].sum().reset_index()
```

```python
yearly_month_wise_sales
```

```python
yearly_month_wise_sales.to_csv('year_month.csv',index=False)
```

```python
# Visualiuze Sales Trends
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# setting up the figuure size
plt.figure(figsize=(18,6))
```

```python
# month-wise sales trnd
plt.subplot(1,3,1)
sns.lineplot(data=month_wise_sales, x='month', y='Sales Channel',marker='o')
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

```python
# year-wise sales trend
plt.subplot(1, 3, 2)
sns.lineplot(data=year_wise_sales, x='year', y='Sales Channel', marker='o')
plt.title('year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total sales')
```

```python
# yearly month-wise sales trend
plt.subplot(1,3,3)
sns.lineplot(data=yearly_month_wise_sales, x='month', y='Sales Channel', hue='year', marker='o')
plt.title('Yealy month-wise sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

```python
plt.tight_layout()
plt.savefig('sales_trends.png')
plt.show()
```
